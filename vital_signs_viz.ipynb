```python
import pandas as pd
import plotly.plotly as py
from plotly.graph_objs import *
red_orange = '#ff3700'
```

```python
df = pd.read_csv('Census_Demographics_2010-2012_2015-03-18.csv',index_col='CSA2010')
list(df.columns)
```

```python
df_white_sorted = df['pwhite10'].apply(lambda x: x[:-1]).astype(float)
df_white_sorted.sort(inplace=True)

fig = plt.figure(figsize=[7,20])
ax1 = plt.subplot(1,1,1)


df_white_sorted.plot(kind='barh', grid='off')

# turn off square border around plot
ax1.spines["top"].set_visible(False)  
ax1.spines["bottom"].set_visible(False)  
ax1.spines["right"].set_visible(False)  
ax1.spines["left"].set_visible(False)

# turn off ticks
ax1.tick_params(axis="both", which="both", bottom="off", top="off",
               labelbottom="on", left="off", right="off", labelleft="on",labelsize=16)

title('% White', fontsize=20)
```

```python
# pandas only

fig = plt.figure(figsize=[8,8])
ax1 = plt.subplot(1,1,1)

#df.drop('Baltimore City', inplace=True)

x = df['pwhite10'].apply(lambda x: x[:-1]).astype(float)
y = df['hhchpov12'].apply(lambda x: x[:-1]).astype(float)
schoolage_pop = df['tpop10'] * df['age18_10'].apply(lambda x: x[:-1]).astype(float)
s = (schoolage_pop/5000.)**2 

plt.Scatter(x,y,
            color=red_orange,
            marker='.',
            s=s,
            alpha=.4)

xlabel('\n% Population White',fontsize=16)
ylabel('% Households w. Children in Poverty\n',fontsize=16)
xlim([-5,100])
ylim([-5,100])
title('Baltimore', fontsize=20)

# turn off square border around plot
ax1.spines["top"].set_visible(False)  
ax1.spines["bottom"].set_visible(False)  
ax1.spines["right"].set_visible(False)  
ax1.spines["left"].set_visible(False)

# turn off ticks
ax1.tick_params(axis="both", which="both", bottom="off", top="off",
               labelbottom="on", left="off", right="off", labelleft="on",labelsize=16)

plt.show()
```

```python
# same plot, interactive from plotly

import plotly.plotly as py

x = df['pwhite10'].apply(lambda x: x[:-1]).astype(float)
y = df['hhchpov12'].apply(lambda x: x[:-1]).astype(float)
schoolage_pop = df['tpop10'] * df['age18_10'].apply(lambda x: x[:-1]).astype(float)


trace1 = Scatter(x=x,
                 y=y,
                 text = np.array(x.index),
                 mode = 'markers',
                 marker = Marker(size = schoolage_pop,
                                 sizemode = 'area',
                                 sizeref = 5e2,
                                 opacity = 0.5,
                                 color = 'blue')
)

layout = Layout(
    title = 'Baltimore: Too Many Kids in Poverty',
    xaxis = XAxis(range = [-5,100],
                  title = '% Population White (2010)',
                  showgrid = False,
                  showline = False,
                  zeroline = False,
                  autotick = False,
                  dtick = 20),
    yaxis = YAxis(range = [-5,100],
                  title = '% Households w. Children in Poverty (2012)',
                  showgrid = False,
                  showline = False,
                  zeroline = False,
                  autotick = False,
                  dtick = 20),
    autosize = False,
    width=500,
    height=500,
    font = Font(size=14),
    hovermode = 'closest'   
)

data = Data([trace1])
fig = Figure(data=data, layout=layout)
py.iplot(fig, filename='Baltimore Percent Child Poverty Percent White')
```
